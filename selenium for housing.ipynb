```python
import csv
import os
import sys
import time

from selenium import webdriver
```

```python
city_name = 'Panghu'
end_page = 4
point = 13-(10-end_page%10)

driver = webdriver.Chrome()
driver.get('https://evertrust.yungching.com.tw/region/%e6%be%8e%e6%b9%96%e7%b8%a3/%e4%b8%8d%e9%99%90/1?t=1,2,3,4&a=&c=&d=6#mainContent')

# Wait the page to fresh.
time.sleep(3)

sys.stderr.write("Write data to csv file...\n")

with open("C:\\Users\yusheng_chen\Desktop\python\Housing price\output\\Housing_"+city_name+".csv", 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(["deal", "status", "address","deal_price","floor_space","single_price","land_space","floor","age","car"])
    
    for page in range(1,end_page+1):
        items = driver.find_elements_by_css_selector("#dealtable tbody tr")
        for item in items:
            tds = item.find_elements_by_css_selector("td")
#             if tds[0].get_attribute("class") != "itemTtitle":
#                 continue
            csvwriter.writerow([tds[0].text, tds[1].text, tds[2].text.encode(sys.stdin.encoding, "replace").decode(sys.stdin.encoding), tds[3].text, tds[4].text, tds[5].text, tds[6].text, tds[7].text, tds[8].text, tds[9].text])
        time.sleep(1.5)
        
        if page == end_page:
            break
        if page == 1:
            nextBtn = driver.find_element_by_xpath('//*[@id="mainContent"]/div[4]/ul/li[11]/a')
        else:
            try:
                nextBtn = driver.find_element_by_xpath('//*[@id="mainContent"]/div[4]/ul/li[13]/a')
            except:
                # handle last page
                nextBtn = driver.find_element_by_xpath('//*[@id="mainContent"]/div[4]/ul/li['+str(point)+']/a')
                
        nextBtn.click()
        
sys.stderr.write("finish~")
csvfile.close()
driver.close()
```

```python
import pandas as pd
import re
import locale
locale.setlocale( locale.LC_ALL, 'en_US.UTF-8' ) 
```

```python
data = pd.read_excel('C:\\Users\yusheng_chen\Desktop\python\Housing price\output\Housing_Tau.xlsx',encoding='utf-8')
data
```

```python
data['address_1'] = data['address'].apply(lambda x:x.split('格局')[0])
data['address_2'] = data['address_1'].apply(lambda x:x.split('\n')[0] if x.count('\n')==1 else x.split('\n')[1])
del data['address_1'] 

def address(add):
    if '段' in add:
        return add.split('段')[0] +'段'
    else:
        return re.sub("[0-9]", "s", add).split('s')[0]
data['地段'] = data['address_2'].apply(address)
def price(p):
    if '\n' in p:
        p = p.split('\n')[0]
    return locale.atoi(p.split('萬')[0])
data['成交價'] = data['deal_price'].apply(price)

def place(p):
    if '\n' in p:
        p = p.split('\n')[0]
    return float(p.split('坪')[0])
data['坪數'] = data['floor_space'].apply(place)
```

```python
data
```

```python
data.groupby('地段').count().sort_values(by=['成交價'],ascending =False).head()
```

```python
import numpy as np
cnt = np.array(data.groupby('地段').count()['deal'])
data2 = data.groupby('地段').sum()
data2['cnt'] = cnt
#data2 = data2.sort_values(by=['成交價'],ascending =False)
avg = np.array(data2['成交價'])/np.array(data2['坪數'])
data2['avg'] = avg
data2.sort_values(by=['avg'],ascending =False)
data2.to_csv('Tau.csv',encoding='utf_8_sig')
```

```python
out = data2[data2['cnt']>=3].sort_values(by=['avg'],ascending =False)
out.to_csv('Kao_top.csv',encoding='utf_8_sig')
```
